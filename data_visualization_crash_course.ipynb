```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# plt.show()
```

```python
x = np.arange(0, 10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, '*')
```

```python
plt.plot(x, y, 'purple')
```

```python
plt.plot(x, y,'go-', label='line 1', linewidth=2 )
```

```python
plt.plot(x, y,'r^--', label='line 1')
```

```python
plt.plot(x, y,'r--', label='line 1')
plt.xlim(0, 5)
plt.ylim(0, 10)
plt.title('Title')
plt.xlabel('X label')
plt.ylabel('Y label')
```

```python
df = pd.read_csv('salaries.csv')
```

```python
df
```

```python
df.plot(x='Salary', y='Age', kind='scatter')
```

```python
df.plot(x='Salary', y='Age', kind='bar')
```

pandas plot types:
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'density' : same as 'kde'
    - 'area' : area plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot
    - 'hexbin' : hexbin plot

```python
df.plot(x='Salary', y='Age', kind='pie')
```
